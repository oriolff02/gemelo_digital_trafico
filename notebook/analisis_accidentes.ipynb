```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Features y target
X = df[["Hora_dia", "Mes_any", "Descripcio_torn_num", "Nom_districte_num", "Descripcio_dia_setmana_num"]]
y = df["accidente"]

# SMOTE para balancear clases
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Modelo XGBoost
model = XGBClassifier(scale_pos_weight=1, use_label_encoder=False, eval_metric='logloss', random_state=42)
model.fit(X_train, y_train)

# Evaluación
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np

# Cargar tu dataframe original de accidentes
df2 = pd.read_csv("dataframe_bueno.csv")

# Número de registros negativos a crear (por ejemplo, igual al 30% de los positivos)
n_negativos = int(len(df2) * 0.3)

# Seleccionar columnas clave
columnas_simular = [
    "Hora_dia", "Mes_any", "Descripcio_torn_num", "Nom_districte_num", 
    "Descripcio_dia_setmana_num", "Codi_barri"
]

# Obtener rangos y valores posibles para cada variable
valores = {}
for col in columnas_simular:
    if df2[col].dtype in [np.int64, np.float64]:
        valores[col] = (df2[col].min(), df2[col].max())
    else:
        valores[col] = df2[col].dropna().unique()

# Crear el DataFrame de negativos
negativos = pd.DataFrame({
    "Hora_dia": np.random.randint(valores["Hora_dia"][0], valores["Hora_dia"][1]+1, size=n_negativos),
    "Mes_any": np.random.randint(valores["Mes_any"][0], valores["Mes_any"][1]+1, size=n_negativos),
    "Descripcio_torn_num": np.random.choice(df2["Descripcio_torn_num"].unique(), size=n_negativos),
    "Nom_districte_num": np.random.choice(df2["Nom_districte_num"].unique(), size=n_negativos),
    "Descripcio_dia_setmana_num": np.random.choice(df2["Descripcio_dia_setmana_num"].unique(), size=n_negativos),
    "Codi_barri": np.random.choice(df2["Codi_barri"].unique(), size=n_negativos)
})

# Rellenar columnas no usadas con valores por defecto o nulos (opcional)
# Aquí puedes añadir otras columnas necesarias, por ejemplo "Numero_expedient" etc.

# Añadir columna de accidente=0
negativos["accidente"] = 0

# Si tu modelo usa otras columnas numéricas, añádelas igual (puedes poner -1 o la moda/media si no importa mucho)

# Concatenar con el DataFrame original
df_final = pd.concat([df2, negativos], ignore_index=True)

# Guardar resultado
df_final.to_csv("dataframe_con_no_accidentes.csv", index=False)

print("✅ Generados y guardados datos negativos en dataframe_con_no_accidentes.csv")
```

```python
import pandas as pd
import numpy as np
from itertools import product

# Carga tus datos
df2 = pd.read_csv("dataframe_bueno.csv")

# Columnas para definir una combinación única
columnas_clave = [
    "Hora_dia", "Mes_any", "Descripcio_torn_num", 
    "Nom_districte_num", "Descripcio_dia_setmana_num", "Codi_barri"
]

# Saca los valores únicos de cada columna
valores_unicos = {col: df2[col].unique() for col in columnas_clave}

# Todas las combinaciones posibles
todas_combis = list(product(*[valores_unicos[col] for col in columnas_clave]))
df_combis = pd.DataFrame(todas_combis, columns=columnas_clave)

# Marca las combinaciones que existen en positivos (accidente=1)
positivos = df2[df2["accidente"] == 1][columnas_clave].drop_duplicates()
# Elimina las combinaciones que ya existen en positivos
negativos = df_combis.merge(positivos, on=columnas_clave, how="left", indicator=True)
negativos = negativos[negativos['_merge'] == 'left_only'].drop('_merge', axis=1)

# Para evitar un dataset gigante, muestrea solo una parte (por ejemplo 30% del número de positivos)
n_negativos = int(len(df2[df2["accidente"] == 1]) * 1.5)
negativos = negativos.sample(n=n_negativos, random_state=42)

# Añade columna 'accidente' = 0
negativos["accidente"] = 0

# Junta con los positivos originales
df_final = pd.concat([df2, negativos], ignore_index=True)

# Guardar resultado
df_final.to_csv("dataframe_con_negativos_ausentes.csv", index=False)

print("✅ Generados y guardados datos negativos AUSENTES en dataframe_con_negativos_ausentes.csv")
```
